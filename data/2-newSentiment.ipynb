```python
from nltk import tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import pandas as pd
from statistics import median, mean
```

```python
df=pd.read_csv('newsData.csv')
data=df.set_index('Date')
```

```python
df.head()
```

```python
sentiment=[]
for i in data.Headlines:
    a=str(i)
    sentences=[]
    compoundSentiment=[]
    lines_list = tokenize.sent_tokenize(a)
    sentences.extend(lines_list)



    sid = SentimentIntensityAnalyzer()
    for sentence in sentences:
        ss = sid.polarity_scores(sentence)
        compoundSentiment.append(float((ss.get('compound'))))
    m=round(mean(compoundSentiment),2)
    sentiment.append(m)
       
```

```python
newsSentiment=pd.DataFrame(sentiment, columns=['newsSentiment'], index=data.index)
```

```python
final_data=pd.read_csv('final_data.csv')
final_data=final_data.set_index('Date', drop=True)
```

```python
final_data=final_data.merge(newsSentiment, left_index=True, right_index=True)
```

```python
final_data.head()
```

```python
#saving to csv
final_data.to_csv('final_data.csv',index=True)
newsSentiment.to_csv('newsSentiment.csv',index=True)
```

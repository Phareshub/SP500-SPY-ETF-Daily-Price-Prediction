```python
import pandas as pd
from pandas import read_csv
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from math import sqrt
import pickle
```

```python
df=pd.read_csv('data/final_data.csv', index_col=0)
```

```python
df.head()
```

```python
target=pd.DataFrame(df.Close_SPY).copy()
```

```python
target.head()
```

```python
target.to_csv('data/target.csv',index=True)
```

```python
for col in df.columns:
    values=df[col].values.reshape(-1,1)
    scaler =MinMaxScaler(feature_range=(0, 1))
    scaler.fit(values)
    df[col]=scaler.transform(values)
    
    with open('scalers/'+str(col)+'_scaler.pickle', 'wb') as t:
        pickle.dump(scaler, t, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
df.head()
```

```python
for col in df.columns:
    values=df[col].values.reshape(-1,1)
    scaler =StandardScaler()
    scaler.fit(values)
    df[col]=scaler.transform(values)
    
    with open('scalers/'+str(col)+'_standardizer.pickle', 'wb') as t:
        pickle.dump(scaler, t, protocol=pickle.HIGHEST_PROTOCOL)
```
